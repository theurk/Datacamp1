# ข้อ 6

```python
x = np.arange(0,7)
y = np.arange(0,7)
x = y**2
y = x**2


plt.title("x = y**2 and y = x**2")
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x,x,c='r',marker='o',label='x = y**2')
plt.plot(x,y,c='b',marker='^',label='y = x**2')

plt.legend(loc=0)

plt.show()
```

```python
x = np.arange(0,7)
y = np.arange(0,7)


plt.title("x = y**2 and y = x**2")
plt.xlabel('x')
plt.ylabel('y')

plt.plot(y**2,y,c='r',marker='o',label='x = y**2')
plt.plot(x,x**2,c='b',marker='^',label='y = x**2')

plt.legend(loc=0)

plt.show()
```
